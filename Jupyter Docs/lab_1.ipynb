# matplotlib module

```python
from matplotlib import pyplot as ptr
t=num.linspace(1,100)
ptr.plot(t,num.sin(t))
```

```python
t=num.linspace(-2*num.pi,2*num.pi,10)
ptr.plot(t,num.cos(2*t))
```

```python
t=num.linspace(-num.pi,num.pi,100)
ptr.plot(t,num.sin(t))
```

# give title to the signal and x,y label to the function

```python
t=num.linspace(-num.pi,num.pi,100)
ptr.plot(t,num.sin(t))
ptr.title("sinusoidal signal of sin")
ptr.xlabel("Time")
ptr.ylabel("Amplitude")
```

# annotate() function 

```python
t=num.linspace(-num.pi,num.pi,100)
ptr.plot(t,num.sin(t))
ptr.title("sinusoidal signal of sin")
ptr.xlabel("Time")
ptr.ylabel("Amplitude")
ptr.annotate("MaxA",xy=(1.5,1))
ptr.annotate("MinA",xy=(-1.5,-1))
```

```python
t=num.linspace(-num.pi,num.pi,100)
ptr.plot(t,num.sin(t),'r+',linewidth=100)
ptr.title("sinusoidal signal of sin")
ptr.xlabel("Time")
ptr.ylabel("Amplitude")
ptr.annotate("MaxA",xy=(1.5,1))
ptr.annotate("MinA",xy=(-1.5,-1))
```

# multiple signals 

```python
t=num.linspace(-num.pi,num.pi,100)
ptr.plot(t,num.sin(t),'r--',linewidth=4)
ptr.plot(t,num.cos(t),'b--',linewidth=4)
ptr.title("sinusoidal signals")
ptr.xlabel("Time")
ptr.ylabel("Amplitude")
ptr.annotate("MaxA",xy=(1.5,1))
ptr.annotate("MinA",xy=(-1.5,-1))
```
